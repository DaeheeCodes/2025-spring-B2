# 2. Weekly graph question

Below are the graphs from last week's weekly graph question.  Suppose you wanted to cause the viewer to examine the last three months of data first.  How could you use preattentive attributes to do that?  Suggest several possibilities.

```python
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
numdates = 12
np.random.seed(0)
time_series = 100 * (np.random.normal(size = numdates) / 5 + np.arange(numdates) / 16)
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Nitrate level (% of eutrophic value)")
```

```python
np.random.seed(1)
time_series = 100 * (1 - (np.random.normal(size = numdates) / 8 + np.arange(numdates) / 16))
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Phosphate level (% of eutrophic value)")
```

# 4. Storytelling With Data plot

Updating the graph from last week, I use difference in hues to grab attention of the reader and point their attention to key dates.

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'insurance_claims_mendeley.csv'
data = pd.read_csv(file_path)

data['incident_date'] = pd.to_datetime(data['incident_date'])

weekly_data = data.groupby(data['incident_date'].dt.to_period('W'))[['injury_claim', 'vehicle_claim']].sum()
weekly_data.reset_index(inplace=True)
weekly_data['incident_date'] = weekly_data['incident_date'].dt.start_time  

plt.figure(figsize=(15, 7))
plt.plot(weekly_data['incident_date'], weekly_data['injury_claim'], marker='o',mec='black',mfc='lime', linestyle='-', label='Injury Claim', color='lightblue')
plt.plot(weekly_data['incident_date'], weekly_data['vehicle_claim'], marker='o',mec='black', mfc='red', linestyle='-', label='Vehicle Claim', color='lightcoral')


plt.title('Weekly Injury and Vehicle Claims Over Time', fontsize=16)
plt.xlabel('Incident Date (Weekly)', fontsize=12)
plt.ylabel('Claim Amount', fontsize=12)
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```
